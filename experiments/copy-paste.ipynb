# Copy-paste detection experiments

### Required imports

```python
import os
import sys

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

import skimage.io
from main.patchmatch import PatchMatch
from main.detector import CopyPasteDetector as CPD
from utils.helpers import detection_results, display_results
```

## First experiment

#### Defining PatchMatch parameteres and running algorithm

```python
img_path = "../images/fleur_copy_paste_normal.png"
img = skimage.io.imread(img_path)
patch_size = 7
iterations = 5
min_norm = 15
diff_threshold = 500.0

detector = CPD(img, patch_size, iterations, min_norm, diff_threshold)
img = detector.img
```

### Defining pairing parameters

```python
cluster_eps=2.0
flat_threshold=0.05
min_cluster_size=300
```

### Running copy-paste pairs detection

```python
paired_result = detector.detect_paired_regions(cluster_eps, flat_threshold, min_cluster_size)
result = detection_results(img, paired_result)
display_results(img, result, diff_threshold)
```
